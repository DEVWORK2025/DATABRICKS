Here the data type comes as by default but we may create different data types based on

pyspark.sql.types

stuct type is used to take the list as a fields 

```
import pyspark.sql.types 
help(pyspark.sql.types )
```

```
from pyspark.sql.types import *


data = [(1,'Nilesh'),(2,'Nita')]
schema1= StructType([StructField(name = 'num', dataType= IntegerType()),
            StructField(name = 'First_name', dataType= StringType())  ])

df = spark.createDataFrame(data = data, schema = schema1)
df.show()
```

```
help(StructType)
help(StructField)
```

```
from pyspark.sql.types import *

schema1 = StructType().add(field= 'id', data_type= IntegerType())\
                        .add(field= 'name', data_type= StringType() )\
                            .add(field= 'gender', data_type= StringType() )\
                                .add(field= 'mobile', data_type= IntegerType() )

# passing list of path with different path
df = spark.read.csv(path=["dbfs:/FileStore/emp1.csv", "dbfs:/FileStore/emp2.csv"], schema = schema1  , header = True)
display(df)
df.printSchema()
```

```
from pyspark.sql.types import *

schema1 = StructType().add(field= 'id', data_type= IntegerType())\
                        .add(field= 'name', data_type= StringType() )\
                            .add(field= 'gender', data_type= StringType() )\
                                .add(field= 'mobile', data_type= IntegerType() )

# passing list of path with different path
df = spark.read.option(key='header', value = True).csv(path =  "dbfs:/FileStore/emp1.csv", schema=schema1 )
display(df)
df.printSchema()
```
