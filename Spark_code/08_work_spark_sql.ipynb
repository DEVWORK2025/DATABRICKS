```
adv_df = spark.read.format('csv').option('header','true').option('inferSchema', 'true')\
    .load('dbfs:/FileStore/Advertising.csv')
adv_df.show()
```

```
adv_df.createOrReplaceTempView("adv_view")
q1 = spark.sql("""select * from adv_view
               where TV > 150""")
q1.show()
```

```
q2 = spark.sql("""select * from adv_view
               where TV > 150""").select('TV','sales')
q2.show()
```

```
q2 = spark.sql("""select * from adv_view
               where TV > 150""").select('TV','sales').distinct()
q2.show()
q2.count()
```
